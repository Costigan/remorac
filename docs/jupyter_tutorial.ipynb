# Remora Jupyter Integration Tutorial

This notebook demonstrates how to use the Remora array language within a Jupyter environment using the `%%remora` magic command.

## 1. Loading the Extension

First, we need to load the `remora.jupyter.magics` extension.

```python
%load_ext remora.jupyter.magics
```

## 2. Basic Execution

You can now use `%%remora` at the start of any cell to execute Remora code. By default, it uses the compiled CPU target.

```python
%%remora
iota 10
```

## 3. Capturing Output into Python Variables

Use the `--out` argument to bind the result of a Remora expression to a Python variable.

```python
%%remora --out my_array
map (* 2.5) (iota 5)
```

```python
print(f"The type of my_array is: {type(my_array)}")
print(f"The contents are: {my_array}")
```

## 4. Visualization with Standard Python Tools

Because Remora returns standard NumPy arrays, you can use any Python visualization library.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(my_array)
plt.title("Visualizing Remora Output")
plt.show()
```

## 5. Using Different Targets

You can choose between `cpu` (default), `interp` (interpreter), and `gpu-nvidia` (if available).

```python
%%remora --target interp
sum (iota 100)
```
